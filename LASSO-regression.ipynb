```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data/Position_Salaries.csv')
```

```python
data
```

```python
x=data.drop(['Salary','Position'],axis=1)
y=data['Salary']
```

```python
x,y
```

```python
from sklearn.linear_model import LinearRegression
LinearReg=LinearRegression()
LinearReg.fit(x,y)
```

```python
lin_pred = LinearReg.predict([[6.5]])  
print(lin_pred)  
```

```python
from sklearn.preprocessing import PolynomialFeatures
Poly_Reg=PolynomialFeatures(degree=2)
x_Poly=Poly_Reg.fit_transform(x)
linearReg_2=LinearRegression()
linearReg_2.fit(x_Poly,y)
```

```python
poly_pred = linearReg_2.predict(Poly_Reg.fit_transform([[6.5]]))  
print(poly_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
Poly_Reg3=PolynomialFeatures(degree=3)
x_Poly3=Poly_Reg3.fit_transform(x)
linearReg_3=LinearRegression()
linearReg_3.fit(x_Poly3,y)
```

```python
poly_pred = linearReg_3.predict(Poly_Reg3.fit_transform([[6.5]]))  
print(poly_pred)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,LinearReg.predict(x),color='red')
plt.title("Bluff detection model(Linear Regression)")  
plt.xlabel("Position Levels")  
plt.ylabel("Salary")  
plt.show()
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,linearReg_2.predict(Poly_Reg.fit_transform(x)),color='red')
plt.title("Bluff detection model(Linear Regression)")  
plt.xlabel("Position Levels")  
plt.ylabel("Salary")  
plt.show()
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,linearReg_3.predict(Poly_Reg3.fit_transform(x)),color='red')
plt.title("Bluff detection model(Linear Regression)")  
plt.xlabel("Position Levels")  
plt.ylabel("Salary")  
plt.show()
```
