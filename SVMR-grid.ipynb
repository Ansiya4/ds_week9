```python
x=data.drop(['MEDV'],axis=1)
y=data['MEDV']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=21,test_size=.25)
```

```python
from sklearn.svm import SVR
svr=SVR()
```

```python
# Define the parameter grid to search
param_grid = {
    'kernel': ['linear', 'rbf'],  # You can also try other kernels
    'C': [0.1, 1, 10, 100],
    'epsilon': [0.01, 0.1, 0.2]
}
```

```python
grid= GridSearchCV(svr, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
# Get the best parameters
best_params = grid.best_params_
print("Best Parameters:", best_params)
```

```python
# Make predictions on the test data using the best model
best_model = grid.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
# Evaluate the best model
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error of the Best Model: {mse}')
```
